## Basic Matplotlib Concepts {#basic-concepts}

Matplotlib is a plotting library for Python that provides a MATLAB-like interface. The main components are:
- **Figure**: The top-level container for all plot elements
- **Axes**: The area where data is plotted
- **Axis**: The number line objects that provide ticks and labels


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set up the plotting style
plt.style.use('default')

```

```python
# With matplotlib, we typically create two-dimensional plots.
# What this means, is we want some data for the x-axis and some data for the y-axis.

x = np.linspace(0, 10, 100)  # 100 evenly spaced numbers from 0 to 10
y = np.sin(x)                 # y is the sine of x

# alternate variable z, random data between 0 and 100
z = np.random.randint(0, 100, size=100)

plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.plot(x, y, "o")  # Plot

plt.figure()
plt.plot(x, z, "o")

# Lineplot with each data point connected
```

```python
# labeling a plot

# For a plot to be informative, we should label it
# Typically, a label on x-axis, y-axis, and a title are the minimum

plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.plot(x, y)
plt.title("Sine Wave")  # Title of the plot
plt.xlabel("X-axis (radians)")  # Label for the x-axis
plt.ylabel("Y-axis (sin(x))")  # Label for the y-axis
```

```python
# labeling a plot

# When we have multiple lines, we should add a legend

c = np.cos(x) 

plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.plot(x, y)
plt.plot(x, c)
plt.title("Sine and Cosine Wave")  # Title of the plot
plt.xlabel("X (radians)")  # Label for the x-axis
plt.ylabel("Y data")  # Label for the y-axis
plt.legend(["sin(x)", "cos(x)"])  # Legend to differentiate the lines
```

### Multiple Lines on One Plot


```python
# Create multiple lines
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)', linewidth=2)
plt.plot(x, y2, label='cos(x)', linewidth=2)
plt.plot(x, y3, label='sin(x)*cos(x)', linewidth=2)
plt.title('Multiple Trigonometric Functions')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Scatter Plots {#scatter-plots}

Scatter plots are useful for visualizing relationships between two variables.


```python
# Generate random data for scatter plot
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5

plt.figure(figsize=(8, 6))

# For scatter plots, we can adjust the transparency with the alpha parameter
# 0 means fully transparent (invisible), 1 means fully opaque (can't see through at all)
plt.scatter(x, y, alpha=0.6, color='blue')
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')

# With plt.grid(), we can add grid lines to the plot
# it is convenient to set the transparency of the grid lines with the alpha parameter often to 0.3
plt.grid(alpha=0.3)
plt.show()
```

```python
# Generate random data for scatter plot
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5
z = -2 * x + np.random.randn(100) * 0.5

plt.figure(figsize=(8, 6))

# For scatter plots, we can adjust the transparency with the alpha parameter
# 0 means fully transparent (invisible), 1 means fully opaque (can't see through at all)
plt.scatter(x, y, alpha=0.6, color='blue')
plt.scatter(x, z, alpha=0.6, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend(['y = 2x + noise', 'z = -2x + noise'])

# With plt.grid(), we can add grid lines to the plot
# it is convenient to set the transparency of the grid lines with the alpha parameter often to 0.3
plt.grid(alpha=0.3)
plt.show()
```

## Histograms {#histograms}

Histograms show the distribution of data and are great for understanding the shape of your data.


```python
# We often use historgrams to visualize distribution of specific trait in our data

# Generate sample data for histogram
np.random.seed(42)
data = np.random.normal(100, 15, 1000)  # Normal distribution

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram Example - Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

### Multiple Histograms

You can also compare multiple distributions in one plot.


```python
# Multiple histograms for comparison
np.random.seed(42)
data1 = np.random.normal(100, 15, 1000)
data2 = np.random.normal(120, 20, 1000)

plt.figure(figsize=(10, 6))
plt.hist(data1, bins=30, alpha=0.7, label='Dataset 1 (=100, =15)', color='blue')
plt.hist(data2, bins=30, alpha=0.7, label='Dataset 2 (=120, =20)', color='red')
plt.title('Multiple Histograms - Comparing Distributions')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Titanic Dataset Visualizations {#titanic-dataset}

Now let's work with the Titanic dataset to create meaningful visualizations.


```python
# Load the Titanic dataset
df = pd.read_csv('titanic.csv')
print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
df
```

```python
gender = df.groupby('Sex')["Fare"].agg(["count", "sum", "mean"])
gender
```

```python
# Groupby (and also iloc) can be a bit frightening at first
# I recommend to play around with it, because they are very useful

survival_by_gender = df.groupby('Sex')['Survived'].agg(['count', 'sum'])
survival_by_gender
```

```python
survival_by_gender['survival_rate'] = survival_by_gender['sum'] / survival_by_gender['count']
survival_by_gender
```

```python
# Survival rate by gender
survival_by_gender = df.groupby('Sex')['Survived'].agg(['count', 'sum'])
survival_by_gender['survival_rate'] = survival_by_gender['sum'] / survival_by_gender['count']

plt.figure(figsize=(8, 6))
bars = plt.bar(survival_by_gender.index, survival_by_gender['survival_rate'], 
               color=['lightblue', 'pink'])
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)

# Add percentage labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.1%}', ha='center', va='bottom')

plt.show()
```

```python
# inspect nulls in age

df["Age"].isnull().sum()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=10, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Survival rate by passenger class
survival_by_class = df.groupby('Pclass')['Survived'].agg(['count', 'sum'])
survival_by_class['survival_rate'] = survival_by_class['sum'] / survival_by_class['count']

plt.figure(figsize=(8, 6))
bars = plt.bar(survival_by_class.index, survival_by_class['survival_rate'], 
               color=['gold', 'silver', 'brown'])
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)

# Add percentage labels
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.1%}', ha='center', va='bottom')

plt.show()
```
